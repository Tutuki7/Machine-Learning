### Fase inicial da aprendizagem não supervisionada

### O objetivo é aplicar métodos de aprendizagem não supervisionada para caracterizar doadores com comportamentos semelhantes, focando apenas quem efetivamente doou e o respetivo DONATION_TYPE.

### Filtraram‑se apenas registos com TARGET_D não nulo (doadores na campanha).
### Criou‑se a variável categórica DONATION_TYPE a partir do montante TARGET_D (já usado na parte supervisionada):

    A: amount ≥ 50
    
    B: 20 ≤ amount < 50
    
    C: 13 ≤ amount < 20
    
    D: 10 ≤ amount < 13
    
    E: amount < 10.

### Para o clustering com KMeans, usou‑se um conjunto alargado de variáveis RFM + contexto socioeconómico (montantes, número de doações, recência, número de promoções, rendimento e valor da casa, etc.), com:
    -Imputação (mediana para numéricas, moda para categóricas).

    -One‑hot encoding nas categóricas, via get_dummies, transformando cada categoria em colunas binárias 0/1.​

    -As features foram normalizadas com StandardScaler e reduzidas a 10 componentes principais (PCs) através de PCA, que explicam cerca de 49,8% da variância total.

    -Clustering com KMeans (escolha de modelo e k)
    - Foram testados modelos Gaussian Mixture (GMM) e KMeans com k ∈ {3, 4, 5, 6} sobre os 10 PCs.
    -A comparação foi feita com métricas internas (silhouette, Davies‑Bouldin, Calinski‑Harabasz) e externas (V‑measure, ARI, FMI) face a DONATION_TYPE.

### O KMeans com k = 4 apresenta:

### As métricas externas para KMeans k=4, comparando com DONATION_TYPE, são baixas:

### Conclusão
    - KMeans com k=4 foi escolhido como modelo final porque apresenta melhores métricas internas do que GMM (silhouette mais alto, Davies‑Bouldin mais baixo, Calinski‑Harabasz muito maior), mantendo desempenho externo similar face a DONATION_TYPE.​
    
    - DONATION_TYPE continua a ser um rótulo fraco como “ground‑truth” para os clusters, mas é útil para mostrar que dentro de cada segmento existem diferentes tipos de montantes nesta campanha, o que pode ajudar a identificar outliers ou oportunidades de up‑selling em subgrupos específicos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score,
    homogeneity_score, completeness_score, v_measure_score,
    adjusted_rand_score, fowlkes_mallows_score
)

# 1. Ler dados e filtrar doadores (TARGET_D não nulo)
df = pd.read_csv("Donors_dataset.csv")
df_donors = df[df["TARGET_D"].notnull()].copy()

def donation_type(amount):
    if amount >= 50:
        return "A"
    elif 20 <= amount < 50:
        return "B"
    elif 13 <= amount < 20:
        return "C"
    elif 10 <= amount < 13:
        return "D"
    else:
        return "E"

df_donors["DONATION_TYPE"] = df_donors["TARGET_D"].apply(donation_type)

# 2. Features focadas (RFM + contexto económico)
numeric_features = [
    "LAST_GIFT_AMT",
    "LIFETIME_GIFT_AMOUNT",
    "LIFETIME_GIFT_COUNT",
    "LIFETIME_MAX_GIFT_AMT",
    "LIFETIME_GIFT_RANGE",
    "MONTHS_SINCE_LAST_GIFT",
    "MONTHS_SINCE_FIRST_GIFT",
    "RECENT_AVG_GIFT_AMT",
    "RECENT_RESPONSE_PROP",
    "NUMBER_PROM_12",
    "LIFETIME_PROM",
    "PER_CAPITA_INCOME",
    "MEDIAN_HOUSEHOLD_INCOME",
    "MEDIAN_HOME_VALUE",
]

categorical_features = [
    "WEALTH_RATING",
    "INCOME_GROUP",
    "SES",
    "URBANICITY",
    "HOME_OWNER",
    "RECENCY_STATUS_96NK",
]

df_clust = df_donors[numeric_features + categorical_features + ["DONATION_TYPE"]].copy()

# 3. Imputação
num_imputer = SimpleImputer(strategy="median")
df_clust[numeric_features] = num_imputer.fit_transform(df_clust[numeric_features])

cat_imputer = SimpleImputer(strategy="most_frequent")
df_clust[categorical_features] = cat_imputer.fit_transform(df_clust[categorical_features])

# 4. One-hot nas categóricas
df_clust_enc = pd.get_dummies(
    df_clust,
    columns=categorical_features,
    drop_first=True
)

X = df_clust_enc.drop("DONATION_TYPE", axis=1)
y_type = df_clust_enc["DONATION_TYPE"]

# 5. Normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. PCA para 10 componentes
pca = PCA(n_components=10, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Variância explicada pelos 10 PCs (Principal Components):", pca.explained_variance_ratio_.sum())

# 7. (Opcional) explorar GMM só para comparação rápida com silhouette
best_k_gmm = None
best_sil_gmm = -1
for k in [3, 4, 5, 6]:
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    labels_k = gmm.fit_predict(X_pca)
    sil = silhouette_score(X_pca, labels_k)
    print(f"GMM k={k}, silhouette={sil:.4f}")
    if sil > best_sil_gmm:
        best_sil_gmm = sil
        best_k_gmm = k

print(f"\nMelhor k (silhouette) para GMM: k={best_k_gmm}, silhouette={best_sil_gmm:.4f}")

# 8. Modelo final: KMeans com k=4 (melhor silhouette nos testes)
k_final = 4
kmeans_final = KMeans(n_clusters=k_final, random_state=42, n_init="auto")
cluster_labels = kmeans_final.fit_predict(X_pca)

# 8a. Métricas internas KMeans k=4 em X_pca
sil_km = silhouette_score(X_pca, cluster_labels)
db_km  = davies_bouldin_score(X_pca, cluster_labels)
ch_km  = calinski_harabasz_score(X_pca, cluster_labels)

print(f"\nMétricas internas KMeans (k={k_final}):")
print(f"Silhouette: {sil_km:.4f}")
print(f"Davies-Bouldin: {db_km:.4f}")       # menor é melhor
print(f"Calinski-Harabasz: {ch_km:.2f}")   # maior é melhor")

# 9. Métricas externas vs DONATION_TYPE (KMeans k=4)
hom = homogeneity_score(y_type, cluster_labels)
comp = completeness_score(y_type, cluster_labels)
v_m  = v_measure_score(y_type, cluster_labels)
ari  = adjusted_rand_score(y_type, cluster_labels)
fmi  = fowlkes_mallows_score(y_type, cluster_labels)

print(f"\nMétricas externas KMeans (k={k_final}):")
print(f"Homogeneity: {hom:.4f}")
print(f"Completeness: {comp:.4f}")
print(f"V-measure: {v_m:.4f}")
print(f"Adjusted Rand Index (ARI): {ari:.4f}")
print(f"Fowlkes-Mallows Index (FMI): {fmi:.4f}")

# 10. Comparação GMM vs KMeans para vários k (mantida para justificar escolha)
print("\n=== Comparação GMM vs KMeans em X_pca ===")

for k in [3, 4, 5, 6]:
    # GMM
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42)
    labels_g = gmm.fit_predict(X_pca)
    sil_g  = silhouette_score(X_pca, labels_g)
    db_g   = davies_bouldin_score(X_pca, labels_g)
    ch_g   = calinski_harabasz_score(X_pca, labels_g)
    v_g    = v_measure_score(y_type, labels_g)
    ari_g  = adjusted_rand_score(y_type, labels_g)
    fmi_g  = fowlkes_mallows_score(y_type, labels_g)

    # KMeans
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels_km = km.fit_predict(X_pca)
    sil_k = silhouette_score(X_pca, labels_km)
    db_k  = davies_bouldin_score(X_pca, labels_km)
    ch_k  = calinski_harabasz_score(X_pca, labels_km)
    v_k   = v_measure_score(y_type, labels_km)
    ari_k = adjusted_rand_score(y_type, labels_km)
    fmi_k = fowlkes_mallows_score(y_type, labels_km)

    print(f"\n--- k={k} ---")
    print(f"GMM   -> Sil: {sil_g:.4f}, DB: {db_g:.4f}, CH: {ch_g:.2f}, V: {v_g:.4f}, ARI: {ari_g:.4f}, FMI: {fmi_g:.4f}")
    print(f"KMeans-> Sil: {sil_k:.4f}, DB: {db_k:.4f}, CH: {ch_k:.2f}, V: {v_k:.4f}, ARI: {ari_k:.4f}, FMI: {fmi_k:.4f}")

# 11. Análise de clusters (tabelas) para o KMeans k=4 escolhido
df_donors_km = df_donors.copy()
df_donors_km["cluster_km"] = cluster_labels

ct = pd.crosstab(df_donors_km["cluster_km"],
                 df_donors_km["DONATION_TYPE"],
                 normalize="index")
print("\nProporção de DONATION_TYPE por cluster (KMeans k=4):")
print(ct)

cluster_profile = df_donors_km.groupby("cluster_km")[
    ["LIFETIME_GIFT_AMOUNT",
     "LIFETIME_GIFT_COUNT",
     "RECENT_AVG_GIFT_AMT",
     "MONTHS_SINCE_LAST_GIFT",
     "MONTHS_SINCE_FIRST_GIFT",
     "WEALTH_RATING",
     "MEDIAN_HOUSEHOLD_INCOME"]
].mean()

print("\nPerfil médio por cluster KMeans (algumas variáveis):")
print(cluster_profile.round(2))

# 12. PCA 2D para visualização (usando X_scaled como antes)
pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_scaled)
print("Variância explicada pelos 2 PCs (plot):",
      pca_2d.explained_variance_ratio_.sum())

plot_df = pd.DataFrame({
    "PC1": X_pca_2d[:, 0],
    "PC2": X_pca_2d[:, 1],
    "cluster_km": cluster_labels,
    "DONATION_TYPE": y_type.values
})

# 12a. Scatter: clusters KMeans em PCA 2D
plt.figure(figsize=(7, 6))
for c in sorted(plot_df["cluster_km"].unique()):
    subset = plot_df[plot_df["cluster_km"] == c]
    plt.scatter(subset["PC1"], subset["PC2"],
                s=10, alpha=0.6, label=f"Cluster {c}")
plt.title("PCA 2D - Clusters KMeans k=4 (doadores)")
plt.xlabel("Eixo X – combinação das variáveis dos doadores (PC1)")
plt.ylabel("Eixo Y – outra combinação das variáveis dos doadores (PC2)")
plt.legend()
plt.tight_layout()
plt.show()

# 12b. Scatter: DONATION_TYPE em PCA 2D
plt.figure(figsize=(7, 6))
for t in sorted(plot_df["DONATION_TYPE"].unique()):
    subset = plot_df[plot_df["DONATION_TYPE"] == t]
    plt.scatter(subset["PC1"], subset["PC2"],
                s=10, alpha=0.6, label=f"Tipo {t}")
plt.title("PCA 2D - DONATION_TYPE (A–E)")
plt.xlabel("Eixo X – combinação das variáveis dos doadores (PC1)")
plt.ylabel("Eixo Y – outra combinação das variáveis dos doadores (PC2)")
plt.legend()
plt.tight_layout()
plt.show()

```

### Esta segunda abordagem segue a lógica da Teórico Prática
    Seleção de um conjunto mais compacto de features RFM + socioeconómicas.
    Normalização (StandardScaler) e PCA para 2 dimensões.
    Clustering com k‑means (k=4) nesse espaço normalizado.
    Treino de um SVM com kernel RBF sobre as coordenadas 2D para aprender fronteiras suaves entre os clusters, e visualização com funções discretescatter e plot2dseparator (como na TP12).

### Resultados Visuais
    - No gráfico "Clusters dos doadores (k-means + SVM em 2D)", os quatro grupos formam blobs bem definidos e o SVM desenha regiões suaves.
    - No gráfico "Donation_Type em 2D (A-E)", usa-se o mesmo espaço 2D, e as classes A-E voltam a aparecer misturadas, sem alinhamento com os regiões dos clusters.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.svm import SVC
import pandas as pd
import numpy as np

df = pd.read_csv("Donors_dataset.csv")

# trabalhar sobre cópia apenas de doadores (como queres)
df_donors = df[df["TARGET_D"].notnull()].copy()

def donation_type(amount):
    if amount >= 50:
        return "A"
    elif 20 <= amount < 50:
        return "B"
    elif 13 <= amount < 20:
        return "C"
    elif 10 <= amount < 13:
        return "D"
    else:
        return "E"

df_donors["DONATION_TYPE"] = df_donors["TARGET_D"].apply(donation_type)
df_donors["DONATION_TYPE"] = df_donors["DONATION_TYPE"].astype("category")

features = [
    "LIFETIME_GIFT_AMOUNT",
    "LIFETIME_GIFT_COUNT",
    "RECENT_AVG_GIFT_AMT",
    "MONTHS_SINCE_LAST_GIFT",
    "MONTHS_SINCE_FIRST_GIFT",
    "WEALTH_RATING",
    "MEDIAN_HOUSEHOLD_INCOME"
]

df_clust = df_donors.dropna(subset=features).copy()

X = df_clust[features].values
y_donation = df_clust["DONATION_TYPE"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- CLUSTERING COM KMEANS, k=4 ---
k = 4  # melhor k pelas métricas internas
kmeans = KMeans(n_clusters=k, random_state=0, n_init="auto")
cluster_labels = kmeans.fit_predict(X_scaled)

df_clust["CLUSTER"] = cluster_labels

# --- PCA 2D para visualização ---
pca_2d = PCA(n_components=2, random_state=0)
X_2d = pca_2d.fit_transform(X_scaled)

# --- SVM em cima dos labels de cluster para aprender as fronteiras ---
svm_clf = SVC(kernel="rbf", C=10, gamma="scale",
              probability=True, random_state=0)
svm_clf.fit(X_2d, cluster_labels)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, colorConverter

def discretescatter(x1, x2, y=None, markers=None, s=10, ax=None,
                    labels=None, padding=.2, alpha=1, c=None,
                    markeredgewidth=None):
    if ax is None:
        ax = plt.gca()
    x1 = np.asarray(x1)
    x2 = np.asarray(x2)
    if y is None:
        y = np.zeros(len(x1), dtype=int)
    y = np.asarray(y)

    uniq = np.unique(y)
    if markers is None:
        markers = ["o", "s", "v", "D", "^", "p", "h", "8", "X", "*"]
    if len(markers) == 1:
        markers = markers * len(uniq)
    if labels is None:
        labels = uniq

    lines = []
    for i, yy in enumerate(uniq):
        mask = (y == yy)
        if c is None:
            color = None
        elif len(c) == 1:
            color = c[0]
        else:
            color = c[i]
        if color is None:
            # usar ciclo de cores do matplotlib
            color = plt.rcParams["axes.prop_cycle"].by_key()["color"][i % len(plt.rcParams["axes.prop_cycle"].by_key()["color"])]
        if markeredgewidth is None:
            mew = 0.5
        else:
            mew = markeredgewidth
        lines.append(
            ax.plot(x1[mask], x2[mask], markers[i],
                    markersize=s, label=labels[i],
                    alpha=alpha, c=color,
                    markeredgewidth=mew,
                    markeredgecolor="black")[0]
        )

    if padding != 0:
        pad1 = x1.std() * padding
        pad2 = x2.std() * padding
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()
        ax.set_xlim(min(x1.min() - pad1, xlim[0]),
                    max(x1.max() + pad1, xlim[1]))
        ax.set_ylim(min(x2.min() - pad2, ylim[0]),
                    max(x2.max() + pad2, ylim[1]))
    return lines

def plot2dseparator(classifier, X, fill=False, ax=None, eps=None,
                    alpha=1, cm=None, linewidth=2, threshold=None,
                    linestyle="solid"):
    if ax is None:
        ax = plt.gca()
    X = np.asarray(X)
    if eps is None:
        eps = X.std() / 2.

    x_min, x_max = X[:, 0].min() - eps, X[:, 0].max() + eps
    y_min, y_max = X[:, 1].min() - eps, X[:, 1].max() + eps

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 500),
        np.linspace(y_min, y_max, 500)
    )
    grid = np.c_[xx.ravel(), yy.ravel()]

    # tentar decision_function; se não existir, usar predict_proba
    try:
        Z = classifier.decision_function(grid)
        if Z.ndim > 1:
            # multi-classe → pegar na classe vencedora
            Z = Z.argmax(axis=1)
        levels = [0] if threshold is None else [threshold]
        if fill:
            if cm is None:
                cm = ListedColormap(["#AAAAFF", "#FFAAAA", "#AAFFAA", "#FFFFAA"])
            ax.contourf(xx, yy, Z.reshape(xx.shape),
                        alpha=alpha, cmap=cm)
        else:
            ax.contour(xx, yy, Z.reshape(xx.shape),
                       levels=levels, colors="k",
                       alpha=alpha, linewidths=linewidth,
                       linestyles=linestyle)
    except AttributeError:
        # fallback para probabilidades (classificação)
        if hasattr(classifier, "predict_proba"):
            proba = classifier.predict_proba(grid)
            if proba.shape[1] == 2:
                Z = proba[:, 1]
                levels = [0.5] if threshold is None else [threshold]
            else:
                Z = proba.argmax(axis=1)
                levels = None
            if fill:
                if cm is None:
                    cm = ListedColormap(["#AAAAFF", "#FFAAAA", "#AAFFAA", "#FFFFAA"])
                ax.contourf(xx, yy, Z.reshape(xx.shape),
                            alpha=alpha, cmap=cm)
            elif levels is not None:
                ax.contour(xx, yy, Z.reshape(xx.shape),
                           levels=levels, colors="k",
                           alpha=alpha, linewidths=linewidth,
                           linestyles=linestyle)

    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plot2dseparator(svm_clf, X_2d, fill=True, alpha=.3)
discretescatter(
    X_2d[:, 0],
    X_2d[:, 1],
    cluster_labels,
    labels=[f"Cluster {c}" for c in np.unique(cluster_labels)]
)
plt.xlabel("Eixo X – combinação das variáveis dos doadores (PC1)")
plt.ylabel("Eixo Y – outra combinação das variáveis dos doadores (PC2)")
plt.title("Clusters dos doadores (KMeans k=4 + SVM em 2D)")
plt.legend(loc="best")
plt.show()

plt.figure(figsize=(6, 5))
discretescatter(
    X_2d[:, 0],
    X_2d[:, 1],
    y_donation,
    labels=[f"Tipo {t}" for t in np.unique(y_donation)]
)
plt.xlabel("Eixo X – combinação das variáveis dos doadores (PC1)")
plt.ylabel("Eixo Y – outra combinação das variáveis dos doadores (PC2)")
plt.title("DONATION_TYPE em 2D (A–E)")
plt.legend(loc="best")
plt.show()
```
